## Data loading

```python
# Load the original data
with open("sipri_graphs.pickle") as f:
    sipri_graphs = cPickle.load(f)
```

```python
EUROPE = ["Albania", "Austria", "Belarus", "Belgium", "Bosnia-Herzegovina",
"Bulgaria", "Croatia", "Czech Republic", "Czechoslovakia", "Denmark", 
"East Germany (GDR)", "Estonia", "Finland", "France", "Germany (FRG)", "Greece", 
"Hungary", "Iceland", "Ireland", "Italy", "Latvia", "Lithuania", "Luxembourg",
"Macedonia", "Netherlands", "Norway", "Poland", "Romania", "Russia", "Serbia",
"Serbia & Mont.", "Slovakia", "Slovenia", "Spain", "Sweden", "Switzerland",
"UK", "UK/PIRA", "Ukraine", "USSR", "Yugoslavia"]
```

```python
# Imports into Europe only with over 20 occurances

edge_count = defaultdict(int)
trade_dict = {}
graph_series = {}
added_edges = {}

for G in sipri_graphs.values():
    for edge in G.edges():
        if edge[1] not in EUROPE: continue
        edge_count[edge] += 1

for y in sipri_graphs.keys():
    G = sipri_graphs[y]
    for edge in G.edges():
        if edge[1] not in EUROPE or edge_count[edge] < 20:
            G.remove_edge(edge[0], edge[1])
            continue
    for node, deg in G.degree().items():
        if deg == 0: G.remove_node(node)
    trade_dict[y] = G.edges()
    graph_series[y] = G
```

```python
# Unpickle previously generated estimate:
with open("probs_n_100.pickle") as f:
    probs = cPickle.load(f)
```

## Running the generative model

```python
conditional_probs = build_conditional_probabilities(probs, 0.25)
```

```python
N = 100
START1 = 1961
END = 2012
real_ensemble = make_graph_ensemble(graph_series[START1-1], conditional_probs, N, START1, END)
null_ensemble = make_null_ensemble(graph_series[START1-1], probs, 0.25, N, START1, END)

mean_degree_real = {}
for year in range(1950, 2012):
    mean_degree_real[year] = np.mean(graph_series[year].degree().values())

fig = plt.figure(figsize=(12,6))
ax = fig.add_subplot(1,1,1)#,axisbg="#eeeeee")
l1, = ax.plot(mean_degree_real.keys(), mean_degree_real.values(), c='b')
for i in range(N):
    vals = {y: np.mean(real_ensemble[y][i].degree().values()) for y in range(START1, END)}
    #vals2 = {y: np.mean(real_ensemble2[y][i].degree().values()) for y in range(1984, 2000)}
    null_vals = {y: np.mean(null_ensemble[y][i].degree().values()) for y in range(START1, END)}
    ax.plot(vals.keys(), vals.values(), c='b', alpha=0.01)
    #plt.plot(vals2.keys(), vals2.values(), c='b', alpha=0.01)
    ax.plot(null_vals.keys(), null_vals.values(), c='r', alpha=0.05)

ax.set_xlim(1950,2011)
ax.set_ylim(0,5)
ax.set_xlabel("Year", fontsize=14)
ax.set_ylabel("Mean Degree", fontsize=14)
l2 = Rectangle((0,0),1,1,fc='b', lw=0, alpha=0.25)
l3 = Rectangle((0,0),1,1,fc='r', lw=0, alpha=0.25)
ax.legend([l1,l2,l3],["Observed", "Generated", "Null"], loc="lower right")
ax.grid(True)
```

```python
def edges_predicted(real_graph, sim_graph):
    real_edges = real_graph.edges()
    sim_edges = sim_graph.edges()
    count = 0.0
    for edge in sim_edges:
        if edge in real_edges: count += 1
        else: count -= 1
    return count/len(sim_edges)
```

```python
for i in range(N):
    vals = {y: edges_predicted(graph_series[y], real_ensemble[y][i]) for y in range(START1, END)}
    null_vals = {y: edges_predicted(graph_series[y], null_ensemble[y][i]) for y in range(START1, END)}
    plt.plot(vals.keys(), vals.values(), c='b', alpha=0.01)
    plt.plot(null_vals.keys(), null_vals.values(), c='r', alpha=0.01)
    
```

```python
## Predict D years forward:
N = 100
START1 = 1951
END = 2012
D = 1

real_ensemble = defaultdict(list)
null_ensemble = defaultdict(list)
for year in range(1950, 2011, D):
    base = graph_series[year]
    for i in range(N):
        #g = build_next_network(base, conditional_probs)
        null_probs = build_null_conditional_probabilities(probs, 0.25)
        last = base
        last_null = base
        for y in range(year+1, year+1+D):
            g = build_next_network(last, conditional_probs)
            n = build_next_network(last_null, null_probs)
            
            real_ensemble[y].append(g)
            null_ensemble[y].append(n)
            last = g
            last_null = n

fig = plt.figure(figsize=(12,6), dpi=300)
ax = fig.add_subplot(1,1,1)#,axisbg="#eeeeee")
l1, = ax.plot(mean_degree_real.keys(), mean_degree_real.values(), c='b')
for i in range(N):
    vals = {y: np.mean(real_ensemble[y][i].degree().values()) for y in range(START1, END)}
    #vals2 = {y: np.mean(real_ensemble2[y][i].degree().values()) for y in range(1984, 2000)}
    null_vals = {y: np.mean(null_ensemble[y][i].degree().values()) for y in range(START1, END)}
    ax.plot(vals.keys(), vals.values(), c='b', alpha=0.01)
    #plt.plot(vals2.keys(), vals2.values(), c='b', alpha=0.01)
    ax.plot(null_vals.keys(), null_vals.values(), c='r', alpha=0.05)

ax.set_xlim(1950,2011)
ax.set_ylim(0,5)
ax.set_xlabel("Year", fontsize=14)
ax.set_ylabel("Mean Degree", fontsize=14)
l2 = Rectangle((0,0),1,1,fc='b', lw=0, alpha=0.25)
l3 = Rectangle((0,0),1,1,fc='r', lw=0, alpha=0.25)
ax.legend([l1,l2,l3],["Observed", "Generated", "Null"], loc="lower right")
ax.grid(True)
```
