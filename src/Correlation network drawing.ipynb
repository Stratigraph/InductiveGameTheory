```python
from collections import defaultdict 

import numpy as np
import networkx as nx
import cPickle
import cProfile


from EventSeries import *
from GenerativeModel import *
```

```python
# Unpickle previously generated estimate:
with open("probs_n_100.pickle") as f:
    probs = cPickle.load(f)
```

```python
G = nx.DiGraph()
for key, p in probs.iteritems():
    if abs(p) > 0.25:
        G.add_edge(key[1], key[0], weight=float(p))
```

```python
nx.write_graphml(G, "prob_net.graphml")
```

```python
loops = 0
for node in G.nodes():
    if (node, node) in G.edges(): loops += 1
print loops
```

```python
G.edge[('Kenya', 'UK')][('Kenya', 'UK')]
```

```python
pos = nx.spring_layout(G)
```

```python
pos_edges = [(u,v) for (u,v,w) in G.edges(data=True) if w['weight'] > 0]
neg_edges = [(u,v) for (u,v,w) in G.edges(data=True) if w['weight'] < 0]

node_colors = []
for node in G.nodes():
    if (node, node) in G.edges():
        w = G.edge[node][node]['weight']
        if w > 0:
            node_colors.append('b')
        else:
            node_colors.append('r')
    else:
        node_colors.append('#666666')

        
node_sizes = [2*d for d in G.degree().values()]
        
fig, ax = plt.subplots(figsize=(12,12))

nx.draw_networkx_nodes(G, pos, ax=ax, node_color=node_colors, node_size=node_sizes, linewidth=0.0)
nx.draw_networkx_edges(G, pos, ax=ax, edgelist=pos_edges, alpha=0.05, edge_color='b')
nx.draw_networkx_edges(G, pos, ax=ax, edgelist=neg_edges, alpha=0.05, edge_color='r', style='--')

# Legend:
ax.set_xlim(-0.2, 1.2)
ax.set_ylim(-0.2, 1.2)

dot = ax.scatter([-5],[-5], c='#666666', s=80, linewidth=0.0)
bdot = ax.scatter([-5],[-5], c='b', s=80, linewidth=0.0)
redline, = ax.plot([-5],[-5], c='r', ls='--')
blueline, = ax.plot([-5,-4],[-5,-4], c='b')
ax.legend([dot, bdot, blueline, redline], 
        ["Dyad - no self-loop", "Dyad - positive self-loop", "Positive correlation", "Negative correlation"], 
        loc="lower right", scatterpoints=1)
ax.axis('off')
```
